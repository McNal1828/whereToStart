```python
# 필요한 라이브러리 import
import pandas as pd
```

```python
# csv 파일로 저장된 raw 데이터 dataframe으로 불러옴 
df_area = pd.read_csv('raw/서울시_상권분석서비스(영역-상권)_epsg_4162.csv', encoding='utf-8', index_col=0)
```

```python
df_area.head(5)
```

```python
# 한글 칼럼명을 가공하기 좋은 형태로 교체하고, 데이터 정규화를 위해 상권_구분, 상권, 자치구, 행정동 칼럼을 분리
df_area.columns = ['cd_div_cd','cd_div_nm','cd_cd','cd_nm','br_cd','br_nm','ab_cd','ab_nm','area','lon','lat']
df_region_info = df_area[['cd_div_cd','cd_div_nm','cd_cd','cd_nm','br_cd','br_nm','ab_cd','ab_nm']]
```

```python
df_region_info.head(5)
```

```python
# 데이터 정규화를 위해 각 코드에 해당하는 한글명 칼럼 제거
df_area.drop(columns=['cd_div_nm','cd_nm','br_nm','ab_nm'], inplace=True)
```

```python
df_area.head(5)
```

```python
# output 폴더에 정제된 데이터 저장
df_area.to_csv('output/region_info.csv', encoding='utf-8')
df_region_info.to_csv('output/cd_area.csv', encoding='utf-8')
```
