```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
```

```python
# csv 파일로 저장된 raw 데이터 df로 불러옴 
df_store = pd.read_csv('raw/서울시_상권분석서비스(점포-상권)_2019년~2023년.csv', encoding='utf-8')
```

```python
df_store.head(5)
```

```python
# 업종 코드와 해당하는 업종명을 unique 메서드를 사용하여 중복 없는 데이터 추출
dict_ = {'st_cd' : list(np.array(df_store['서비스_업종_코드'].unique())),
        'st_nm' : list(np.array(df_store['서비스_업종_코드_명'].unique()))}
```

```python
# dataframe 형식으로 변환
df_code = pd.DataFrame.from_dict(dict_)
df_code
```

```python
# 데이터 정규화와 칼럼명 교체
df_store.drop(columns=['상권_구분_코드_명','상권_코드_명','서비스_업종_코드_명'], inplace=True)
df_store.columns = ['aq_cd','cd_div_cd','cd_cd','st_cd','total_ns','si_ns','total_or','open_ns','total_cr','close_ns','franchise_ns']
```

```python
df_store.head(5)
```

```python
# output 폴더에 저장
df_store.to_csv('output/cd_store.csv', encoding='utf-8')
df_code.to_csv('output/code_info.csv', encoding='utf-8')
```
