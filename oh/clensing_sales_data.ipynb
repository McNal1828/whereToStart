```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
```

```python
# csv 파일로 저장된 raw 데이터 df로 불러옴 
df_sales = pd.read_csv('raw/서울시_상권분석서비스(추정매출-상권)_2019년~2023년.csv', encoding='utf-8')
```

```python
# 데이터 정제를 위해 칼럼 간소화
df_sales.drop(columns=['상권_구분_코드_명','상권_코드_명','서비스_업종_코드_명'], inplace=True)
```

```python
# 상권 코드와 업종코드의 총 개수 확인
print(df_sales['상권_코드'].value_counts().count(), df_sales['서비스_업종_코드'].value_counts().count())
```

```python
# 해당 상권에서 한 번이라도 존재했던 업종을 모든 분기에서 출력되도록 수정
arr = np.array(df_sales['상권_코드'])
quaters = [20191, 20192, 20193, 20194, 20201, 20202, 20203, 20204, 20211, 20212, 20213, 20214, 20221, 20222, 20223, 20224, 20231, 20232, 20233]

dict_ = {i:np.array(df_sales[df_sales['상권_코드'] == i]['서비스_업종_코드'].value_counts().index) for i in arr}
```

```python
# 데이터 정규화 작업 및 칼럼명 교체
df_sales.drop(columns=['상권_구분_코드_명','상권_코드_명','서비스_업종_코드_명'], inplace=True)
df_sales.columns = ['cd_cd','st_cd','aq_cd','cd_div_cd','cm_sa','cm_sc',
               'wd_sa','we_sa','mon_sa','tue_sa','wed_sa','thu_sa','fri_sa',
               'sat_sa','sun_sa','00_06_sa','06_11_sa','11_14_sa','14_17_sa','17_21_sa','21_24_sa','man_sa','woman_sa','10s_sa',
               '20s_sa','30s_sa','40s_sa','50s_sa','over_60s_sa',
               'wd_sc','we_sc','mon_sc','tue_sc','wed_sc','thu_sc','fri_sc',
               'sat_sc','sun_sc','00_06_sc','06_11_sc','11_14_sc','14_17_sc','17_21_sc','21_24_sc','man_sc','woman_sc','10s_sc',
               '20s_sc','30s_sc','40s_sc','50s_sc','over_60s_sc']
```
